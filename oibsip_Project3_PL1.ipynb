```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/SAMRAJ/Downloads/archive (1)/AB_NYC_2019.csv')
df
```

```python
# Step 1: Handling Missing Values
# Fill missing values in 'last_review' with a placeholder date 'No Review'
df['last_review'].fillna('No Review')
df
```

```python
# Fill missing values in 'reviews_per_month' with 0
df['reviews_per_month'].fillna(0)
df
```

```python
# Drop rows with missing values in 'host_name' and 'name'
df.dropna(subset=['host_name', 'name'])
df
```

```python
# Step 2: Ensure Correct Data Types
# Convert 'last_review' to datetime, if it's not 'No Review'
df['last_review'] = df['last_review'].apply(lambda x: pd.to_datetime(x) if x != 'No Review' else x)
df
```

```python
# Step 3: Remove Duplicate Entries
df.drop_duplicates()
```

```python
# Step 4: Addressing Potential Outliers
# Define a function to detect outliers using IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Detect outliers in the 'price' column
outliers_price = detect_outliers(df, 'price')
print(f"Number of outliers in 'price': {len(outliers_price)}")

# Optionally, remove outliers
df = df[~df.index.isin(outliers_price.index)]
```

```python
df
```

```python
df.to_csv('AB_NYC_2019_cleaned.csv', index=False)
print("Data cleaning complete. Cleaned data saved to 'AB_NYC_2019_cleaned.csv'.")
```
